# Less feature part

```python
from sklearn import datasets, svm, metrics, tree
from sklearn.preprocessing import StandardScaler

import numpy as np
import pandas as pd
from itertools import cycle

dtype={'population':np.int64,
        'NofCrime' : np.int64,
        'Nof5CategoryCrime' : np.int64,
        'NofOtherCrime' :  np.int64,
        'midnightCrime' :  np.int64,
        'morningCrime' :  np.int64,
        'dayCrime': np.int64,
        'nightCrime' :  np.int64,
        'bad_health' :  np.float64,
        'bad_health_range' :  np.int64,
        'bad_health_range10' :  np.int64
      }
data = pd.read_csv("Data_NofCrime.csv",dtype=dtype, usecols=dtype)
#print(data)
drop_list = [
        'population',
        #'NofCrime',
        #'Nof5CategoryCrime',
        #'NofOtherCrime',
        #'midnightCrime',
        'morningCrime',
        'dayCrime',
        #'nightCrime',
        'bad_health',
        #'bad_health_range',
        'bad_health_range10'
    ]

range_data_7 = data.drop(drop_list, axis=1)

Train_X_7 = range_data_7.drop('bad_health_range',axis = 1)
scaler.fit(Train_X_7)
Train_X_7 = scaler.transform(Train_X_7)
Class_Y_7 = range_data_7['bad_health_range']
```

```python
from sklearn import datasets, svm, metrics, tree
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.neural_network import MLPClassifier

X_train, X_test, y_train, y_test = train_test_split(
    Train_X_7, Class_Y_7, test_size=0.1, random_state=0)


forest = RandomForestClassifier(n_estimators=31,max_depth=8)
en_list = []

for x in range(50):
    scores = cross_val_score(forest, X_train, y_train, cv=10)
    en_list.append(["Ensemble Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2), scores.mean(), scores.std() * 2])
    
from sklearn import svm
svm_list = []
svm_basic = svm.SVC(kernel='rbf')
for x in range(50):
    scores = cross_val_score(svm_basic, X_train, y_train, cv=10)
    svm_list.append(["SVM Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2), scores.mean(), scores.std() * 2])
    

tree_list = []
clf_tree = tree.DecisionTreeClassifier()
clf_tree = clf_tree.fit(X_train, y_train)
for x in range(50):
    scores = cross_val_score(clf_tree, X_train, y_train, cv=10)
    tree_list.append(["DecisionTree Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2), scores.mean(), scores.std() * 2])
    

mlp = MLPClassifier(hidden_layer_sizes=(15,15,15),max_iter=1500)
nn_list = []

for x in range(50):
    scores = cross_val_score(mlp, X_train, y_train, cv=10)
    nn_list.append(["NeuralNetwork Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2), scores.mean(), scores.std() * 2])
    
    
import pylab as pl

pl.clf()
pl.close('all')
fig, ax = pl.subplots(figsize=(8, 8))

for i in range(50):    
    c1 = ax.scatter(1, tree_list[i][1], c='r', marker = '+')
    
for i in range(50):    
    c2 = ax.scatter(2, svm_list[i][1], c='g', marker = 'o')
    
for i in range(50):    
    c3 = ax.scatter(3, nn_list[i][1], c='b', marker = '*')
    
for i in range(50):    
    c4 = ax.scatter(4, en_list[i][1], c='gold', marker = 'h')
        
ax.legend([c1, c2, c3, c4], ['DecisionTree ', 'SVM',   'NeuralNetwork', 'RandomForest'], fontsize=20)
pl.title('50 times classification comparison', fontsize=30)


labels = [item.get_text() for item in ax.get_xticklabels()]
empty_string_labels = ['']*len(labels)
ax.set_yticks(list(np.arange(0.30, 0.55, 0.01)))
ax.set_xticklabels(empty_string_labels)
#ax.set_yticklabels(labelsize=30)
#figure(figsize=(20,10))
pl.ylabel("Accuracy", fontsize=25 )
avg_tree = sum([x[1] for x in tree_list]) / 50
avg_svm = sum([x[1] for x in svm_list]) / 50
avg_nn = sum([x[1] for x in nn_list]) / 50
avg_en = sum([x[1] for x in en_list]) / 50

pl.axhline(avg_tree, 0.005, 0.095, c='black')
pl.axhline(avg_svm, 0.305, 0.395, c='black')
pl.axhline(avg_nn, 0.605, 0.695, c='black')
pl.axhline(avg_en, 0.905, 0.995, c='black')

#pl.savefig('/home/junlinux/Desktop/INF552/HW3/Final_data/50tiems_cls_withCrimeTme_upload.pdf')
pl.show()
```
